```python
import pandas as pd
```

```python
xls = pd.ExcelFile('balanco_marisa.xls')
```

```python
dre = pd.read_excel(xls, 'Dem. Result.', header=1)
dre
```

```python
bp = pd.read_excel(xls, 'Bal. Patrim.', header=1)
bp
```

```python
dre = dre.transpose()
pd.set_option('display.max_columns', None)
dre.head()
```

```python
bp = bp.transpose()
pd.set_option('display.max_columns', None)
bp.head()
```

```python
bpr = bp.iloc[:,[4,5,27]]
bpr.head()
```

```python
bpr = bpr.rename(columns={4:'clientes',5:'estoques',27:'fornecedores'})
bpr.head()
```

```python
bpr = bpr.drop(bpr.index[[0]])
bpr.head()
```

```python
drer = dre.iloc[:,[2,3]]
drer.head()
```

```python
drer = drer.rename(columns={2:'receita_liquida',3:'custo'})
drer.head()
```

```python
drer = drer.drop(drer.index[[0]])
drer.head()
```

```python
print(drer['receita_liquida'].isnull().sum())
print(drer['custo'].isnull().sum())
```

```python
bpr.head(1)
```

```python
print(bpr['clientes'].isnull().sum())
print(bpr['estoques'].isnull().sum())
print(bpr['fornecedores'].isnull().sum())
```

```python
pme = abs((bpr.estoques/drer.custo)*90)
pme
```

```python
pmr = abs((bpr.clientes/drer.receita_liquida)*90)
pmr
```

```python
pmp = abs((bpr.fornecedores/drer.custo)*90)
pmp
```

```python
co = pme+pmr
co
```

```python
cc = co - pmp
cc
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(35,12))
ax = fig.add_subplot(111) 
plt.xticks(rotation=90) # Rotação do eixo x do grafico
plt.title('Prazos Médios Marisa', size = 16)
plt.plot(pmp, "red")
plt.plot(pmr, "blue")
plt.plot(pme, "green")
arrowprops=dict(arrowstyle='->', color='red', linewidth=5, mutation_scale=15)
ax.annotate('Interrupção das Vendas causadas pelo Covid-19', xy = ('31/03/2020', 150), fontsize=15, color='red',
            xytext=(12, 150), arrowprops=arrowprops)
plt.legend(['Prazo Médio de Pagamento', 'Prazo Médio de Recebimento', 'Prazo Médio de Estoque'], loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(24,8))
plt.xticks(rotation=90) # Rotação do eixo x do grafico
plt.title('Ciclo de Caixa vs Ciclo Operacional', size = 16)
plt.plot(cc, '--', color='red')
plt.plot(co, '--', color='blue')
plt.legend(['Ciclo Caixa', 'Ciclo Operacional'], loc='upper left')
plt.show()
```
