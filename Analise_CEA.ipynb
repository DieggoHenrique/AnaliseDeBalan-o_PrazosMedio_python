```python
# Importando as bibliotecas para modelagem de dados
import pandas as pd
```

```python
# Carregando excel da base de dados
# Fonte: balanço e DRE site Fundamentos
xls = pd.ExcelFile('balanco_cea.xls')
```

```python
# Carregando a DRE da base de dados
dre = pd.read_excel(xls, 'Dem. Result.', header=1)
dre.head().T
```

```python
# Carregando o balanço Patrimonial
bp = pd.read_excel(xls, 'Bal. Patrim.', header=1)
bp.head().T
```

```python
# Transpondo a base de dados
dre = dre.transpose()
# Comando para mostrar todas as colunas
pd.set_option('display.max_columns', None)
dre.head()
```

```python
# Transpondo a base de dados
bp = bp.transpose()
# Comando para mostrar todas as colunas
pd.set_option('display.max_columns', None)
bp.head()
```

```python
# Separando as variáveis que irão compor os cálculos
bpr = bp.iloc[:,[4,5,27]]
bpr.head()
```

```python
# Renomeando o nome das colunas
bpr = bpr.rename(columns={4:'clientes',5:'estoques',27:'fornecedores'})
bpr.head()
```

```python
# Removendo a linha do título
bpr = bpr.drop(bpr.index[[0]])
bpr.head()
```

```python
drer = dre.iloc[:,[2,3]]
drer.head()
```

```python
drer = drer.rename(columns={2:'receita_liquida',3:'custo'})
drer.head()
```

```python
drer = drer.drop(drer.index[[0]])
drer.head()
```

```python
# Verificando valores zerados
print(drer.isnull().sum())
```

```python
# Removendo linhas com valores zerados
index_para_dropar = drer.loc[drer['receita_liquida'] == 0,['receita_liquida', 'custo']].index
```

```python
# Verificando o index
index_para_dropar
```

```python
drer.drop(index_para_dropar, axis=0, inplace=True)
drer.drop(drer.index[[-1]], inplace=True)
```

```python
print(bpr.isnull().sum())
```

```python
bpr.drop(index_para_dropar, inplace=True)
bpr.drop(bpr.index[[-1]], inplace=True)
```

```python
pme = abs((bpr.estoques/drer.custo)*90)
pme
```

```python
pmr = abs((bpr.clientes/drer.receita_liquida)*90)
pmr
```

```python
pmp = abs((bpr.fornecedores/drer.custo)*90)
pmp
```

```python
co = pme+pmr
co
```

```python
cc = co - pmp
cc
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig = plt.figure(figsize=(24,8))
ax = fig.add_subplot(111) 
plt.title('Prazos Médios C&A', size = 16)
plt.plot(pmp, "red")
plt.plot(pmr, "blue")
plt.plot(pme, "green")
arrowprops=dict(arrowstyle='->', color='red', linewidth=5, mutation_scale=15)
ax.annotate('Interrupção das Vendas causadas pelo Covid-19', xy = ('30/06/2020', 400), fontsize=15, color='red',
            xytext=(2, 400), arrowprops=arrowprops)
plt.legend(['Prazo Médio de Pagamento', 'Prazo Médio de Recebimento', 'Prazo Médio de Estoque'], loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(24,8))
plt.title('Ciclo de Caixa vs Ciclo Operacional', size = 16)
plt.plot(cc, '--', color='red')
plt.plot(co, '--', color='blue')
plt.legend(['Ciclo Caixa', 'Ciclo Operacional'], loc='upper left')
plt.show()
```
